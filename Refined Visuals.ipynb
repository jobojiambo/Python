```python
import pandas as pd
my_data = pd.read_csv('happyscore_income.csv')
import matplotlib.pyplot as plt
my_data
```

```python
my_data.sort_values('median_income')
my_data.sort_values('GDP')
my_data
middle_income = my_data[my_data['median_income']>5000] 
middle_income
import matplotlib.pyplot as plt
plt.xlabel('GDP')
plt.ylabel('median_income')
plt.scatter(middle_income['GDP'],middle_income['median_income'])
plt.text(middle_income.iloc[-1]['GDP'],
        middle_income.iloc[-1]['median_income'],
        middle_income.iloc[-1]['country']
        )
plt.text(middle_income.iloc[0]['GDP'],
        middle_income.iloc[0]['median_income'],
        middle_income.iloc[0]['country']
        )

```
