```python
import matplotlib.pyplot as plt
income = data['avg_income']
happy = data['happyScore']
inq = data['income_inequality']
plt.xlabel('income')
plt.ylabel('inq')
plt.scatter(income, happy, s = inq*10, alpha=0.35)
```

```python
from sklearn.cluster import KMeans
import numpy as np
income_happy = np.column_stack((income, happy))
income_happy
km_results = KMeans(n_clusters=2).fit(income_happy)
cluster = km_results.cluster_centers_
plt.scatter(income,happy)
plt.scatter(cluster[:,0], cluster[:,1], s=1055)
```
