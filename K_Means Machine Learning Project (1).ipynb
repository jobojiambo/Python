```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Banknote = pd.read_csv('Banknote_Project')
Banknote
```

```python
mean = np.mean(Banknote)
variance = np.var(Banknote)
Deviation = np.std(Banknote)
print([mean, variance, Deviation])
```

```python
Banknote.sort_values('V2', inplace = True)
Banknote
Cluster_1 = Banknote[ Banknote['V2'] < 3]
Cluster_1
mean_1 = np.mean(Cluster_1)
mean_1

plt.scatter(Cluster_1['V2'], Cluster_1['V1'], alpha = 0.5)
```

```python
Banknote.sort_values('V2', inplace = True)
Banknote
Cluster_2 = Banknote[ Banknote['V2'] >= 3]
Cluster_2

mean_2 = np.mean(Cluster_2)
mean_2
plt.scatter(Cluster_2['V2'], Cluster_2['V1'], alpha=0.5)
```

```python
from sklearn.cluster import KMeans
C1 = Cluster_1['V1']
C2 = Cluster_1['V2']
C3 = Cluster_2['V1']
C4 = Cluster_2['V2']
[C1,C2]
[C3,C4]
```

```python
Range_V1 = np.column_stack((C1,C2))
km_Res = KMeans(n_clusters = 4).fit(Range_V1)
print(Range_V1)
```

```python
Cluster_Centers = km_Res.cluster_centers_
Cluster_Centers
plt.scatter(C1,C2, alpha = 0.75)
plt.scatter(C3,C4, alpha = 0.75)
plt.scatter(Cluster_Centers[:,0], Cluster_Centers[:,1], alpha = 0.75, s = 1000)
```

```python
y_predicted = km_Res.fit_predict(Banknote[['V1', 'V2']])
y_predicted
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
Banknote['New_Col'] = y_predicted
Banknote.head()
DF1 = Banknote[Banknote.New_Col==0]
DF2 = Banknote[Banknote.New_Col==1]
DF3 = Banknote[Banknote.New_Col==2]
DF4 = Banknote[Banknote.New_Col==3]
DF5 = Banknote[Banknote.New_Col==4]

plt.scatter(DF1.V1,DF1['V2'])
plt.scatter(DF2.V1,DF2['V2'])
plt.scatter(DF3.V1,DF3['V2'])
plt.scatter(DF4.V1,DF4['V2'])

plt.scatter(Cluster_Centers[:,0], Cluster_Centers[:,1], alpha = 0.5, s = 1000)
plt.xlabel('V1')
plt.ylabel('V2')
```

```python
scaler = MinMaxScaler()
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(Banknote[['V1','V2']])
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('k_entries')
plt.ylabel('Sum of Square Errors')
plt.scatter(k_rng,sse)
```
